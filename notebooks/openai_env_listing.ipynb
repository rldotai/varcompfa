# Environments Available

Here's a quick way to check what environments are available and can be made.

Also, it looks like some of these import TensorFlow? 
What's that about?

```python
import gym 
import numpy as np 
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
TAB = '\t'
envlst = sorted(gym.envs.registry.all(), key=lambda x: x.id)
errlst = []
for i in envlst:
    try:
        env = i.make()
        print(i.id)
        print(TAB, 'obs:', env.observation_space)
        print(TAB, 'act:', env.action_space)
    except Exception as e:
#         print(i.id)
#         print(TAB, '\x1b[31m Cannot instantiate environment \x1b[0m', flush=True)
        errlst.append(i)

print('*'*60)
print("Cannot make:")
print('*'*60)
for i in errlst:
    print(i.id)
```

# Where Are The Environments That Can't Be Instantiated?

```python
for i in errlst:
    print(i.id)
    print(TAB, i._entry_point)
```

# What Was The Exception Raised?

```python
for i in errlst:
    print(i.id)
    try:
        env = i.make()
    except Exception as e:
        print(e)
        
    print('*'*80)
```
